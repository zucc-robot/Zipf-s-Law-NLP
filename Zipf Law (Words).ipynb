```python
import re
import numpy as np
import matplotlib.pyplot as plt
from operator import itemgetter 
from matplotlib.pyplot import figure
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
frequency = {}
open_file = open('Tom Sawyer.txt', 'r', encoding = "utf8")
file_to_string = open_file.read()
file_to_string = file_to_string.lower()
words = re.findall(r'(\b[A-Za-z][a-z]{2,9}\b)', file_to_string)
token_ = 0
type_ = 0
xlog = []
ylog = []
for word in words:
    count = frequency.get(word,0)
    frequency[word] = count + 1
    token_=token_+1
     
for key, value in reversed(sorted(frequency.items(), key = itemgetter(1))):
    type_=type_+1
    print (type_, key, value)
    xlog.append(np.log10(type_))
    ylog.append(np.log10(value))
    plt.plot(ylog, xlog)
plt.xlabel('Rank') 
plt.ylabel('Frequency') 
plt.title('Zipf Law') 
plt.show() 
ttr = type_/token_
ttr_inverse = 1/ttr
```

Analysis of the 5 Most Common Words:

1)The:  Definite Article: Contains two out of the five most commonly used letters.
3)And:  Conjunction: Contains one of the five most commonly used letters
3)Was:  Verb: Contains one of the five most commonly used letters 
4)That: Pronoun/Adjective/Adverb: Contains three cumulative occurances of two most commonly used letters 
5)You:  Possessive/Objective Pronoun: Contains one of the 5 most commonly used letters

P.S: 3 out of the 5 most common letters are vowels. A disproportionately high fraction considering that vowels are only 19.2% of the alphabet, whereas consonants, that form 80.7% of the alphabet, occur only twice in the 5 most common letters.
